```python
# Import math module and all lists from separate notebook.
import math
%run 'Midjourney Function - 512 Bits - Lists.ipynb'

# Master list containing all nested lists above.
lists = [stylize, seed, detail, styles, weights, red, green, blue, weights, adjectives1, weights, adjectives2, weights, characters, weights, adverbs1, weights, adverbs2, weights, verbs, weights, prepositions, weights, red, green, blue, weights, adjectives1, weights, adjectives2, weights, objects, weights, adverbs1, weights, verbs, weights, red, green, blue, weights, adjectives1, weights, adjectives2, weights, objects, weights, emotions, weights, locations, weights, time_periods, weights, artists, weights]

# Master list containing titles of all nested lists above. Serves iterative purpose.
listsNames = ["Stylize", "Seed", "Detail", "Styles", "Weight", "Red", "Green", "Blue", "Weight", "Adjectives", "Weight", "Adjectives", "Weight", "Characters", "Weight", "Adverbs", "Weight", "Adverbs", "Weight", "Verbs", "Weight", "Prepositions", "Weight", "Red", "Green", "Blue", "Weight", "Adjectives", "Weight", "Adjectives", "Weight", "Objects", "Weight", "Adverbs", "Weight", "Verbs", "Weight", "Red", "Green", "Blue", "Weight", "Adjectives", "Weight", "Adjectives", "Weight", "Objects", "Weight", "Emotions", "Weight", "Locations", "Weight", "Time Periods", "Weight", "Artists", "Weight"]

# Prompting user to input public key and hint.
pubkey = input("Insert pubkey. ")
hint = input("Insert hint. ")

# Removing '0x' from beginning of public key and hint.
pubkey = pubkey[2:]
hint = hint[2:]

# Converting public key from hexadecimal to integer, then to binary form. Removing '0b' from beginning of binary public key and storing the following 257 bits. Same process is performed for the hint, however, the entirety of the hint is maintained, excluding '0b' at start (255 bits). Variable 'result' holds these appended strings (512 bits).
def convertToBinary(pubkey, hint):
    result = ""
    pubkeybin = bin(int(pubkey, 16))
    pubkeybin = pubkeybin[2:259]
    hintbin = bin(int(hint, 16))
    hintbin = hintbin[2:]
    result += pubkeybin + hintbin
    return result

# Call binary conversion function for the inputted public key and hint.
bitString = convertToBinary(pubkey, hint)

results = []

# Iterate through binary 'result' from start value to index log base two of length of particular list in master list. Convert to an integer value. Append the value stored at this integer index within the particular list to an empty list 'results'. Append only weight values to empty list 'sum'. 
def getValues(start, end, i, decimal, results, sum):
    for el in bitString[start:end]:
        decimal = decimal*2 + int(el)
    results.append(lists[i][decimal])
    if listsNames[i] == "Weight":
      sum.append(float(lists[i][decimal]))

# First start value labelled index 0. First end value labelled index log base two of length of first list in master list. 
start = 0
i = 0
end = int(math.log2(len(lists[i])))
decimal = 0
sum = []

# Iterate through master list of nested lists, executing 'getValues' function and changing start value and end value upon each iteration. Start value becomes previous end value; end value becomes index log base two of length of following list in master list. Variable 'decimal' holding integer value is reset to zero.
while i < len(listsNames):
    getValues(start, end, i, decimal, results, sum)
    if i == len(listsNames) - 1:
      break
    else:
      start += int(math.log2(len(lists[i])))
      end += int(math.log2(len(lists[i + 1])))
      i += 1
      decimal = 0 

# Add all weight values yielded to variable 'test'. 
test = 0
for value in sum:
  test += value

# If sum of weights is positive, function proceeds to print values in a Midjourney-compatible fashion. If sum of weights is negative, Midjourney prompt will not proceed. In that case, all weights are flipped to be positive and values are printed in the same manner.
# A [detail][styles] with [colours][adjectives][adjectives][characters][adverbs][adverbs][verbs][prepositions][colours][adjectives][adjectives][objects][adverbs][verbs] and [colours][adjectives][adjectives][objects] in [emotions][locations][time period] in style of [artists] --stylize x -- seed y.
if test > 0:
  print(results[2] + " " + results[3] + "::" + results[4] + " " + results[5] + results[6] + results[7] + "::" + results[8] + " " + results[9] + "::" + results[10] + " " + results[11] + "::" + results[12] + " " + results[13] + "::" + results[14] + " " + results[15] + "::" + results[16] + " " + results[17] + "::" + results[18] + " " + results[19] + "::" + results[20] + " " + results[21] + "::" + results[22] + " " + results[23] + results[24] + results[25] + "::" + results[26] + " " + results[27] + "::" + results[28] + " " + results[29] + "::" + results[30] + " " + results[31] + "::" + results[32] + " " + results[33] + "::" + results[34] + " " + results[35] + "::" + results[36] + " " + results[37] + results[38] + results[39] + "::" + results[40] + " " + results[41] + "::" + results[42] + " " + results[43] + "::" + results[44] + " " + results[45] + "::" + results[46] + " " + results[47] + "::" + results[48] + " " + results[49] + "::" + results[50] + " " + results[51] + "::" + results[52] + " " + results[53] + "::" + results[54] + " " + results[0] + " " + results[1])
  
else:
  new_list = []
  new_list = [el.replace("-", "") for el in results[2:]]
  new_list = [results[0]] + [results[1]] + new_list
  print(new_list[2] + " " + new_list[3] + "::" + new_list[4] + " " + new_list[5] + new_list[6] + new_list[7] + "::" + new_list[8] + " " + new_list[9] + "::" + new_list[10] + " " + new_list[11] + "::" + new_list[12] + " " + new_list[13] + "::" + new_list[14] + " " + new_list[15] + "::" + new_list[16] + " " + new_list[17] + "::" + new_list[18] + " " + new_list[19] + "::" + new_list[20] + " " + new_list[21] + "::" + new_list[22] + " " + new_list[23] + new_list[24] + new_list[25] + "::" + new_list[26] + " " + new_list[27] + "::" + new_list[28] + " " + new_list[29] + "::" + new_list[30] + " " + new_list[31] + "::" + new_list[32] + " " + new_list[33] + "::" + new_list[34] + " " + new_list[35] + "::" + new_list[36] + " " + new_list[37] + new_list[38] + new_list[39] + "::" + new_list[40] + " " + new_list[41] + "::" + new_list[42] + " " + new_list[43] + "::" + new_list[44] + " " + new_list[45] + "::" + new_list[46] + " " + new_list[47] + "::" + new_list[48] + " " + new_list[49] + "::" + new_list[50] + " " + new_list[51] + "::" + new_list[52] + " " + new_list[53] + "::" + new_list[54] + " " + new_list[0] + " " + new_list[1])
```
