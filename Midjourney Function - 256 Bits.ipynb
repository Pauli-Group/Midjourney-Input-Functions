```python
# Import math module and all lists from separate notebook.
import math
%run 'Midjourney Function - 256 Bits - Lists.ipynb'

# Master list containing all nested lists above.
lists = [stylize, detail, styles, weights, red, green, blue, weights, adjectives1, weights, adjectives2, weights, characters, weights, adverbs1, weights, adverbs2, weights, verbs, weights, prepositions, weights, adjectives1, weights, objects, weights, locations, weights]

# Master list containing titles of all nested lists above. Serves a purpose for iteration only.
listsNames = ["Stylize", "Detail", "Styles", "Weights", "Red", "Green", "Blue", "Weights", "Adjectives1", "Weights", "Adjectives2", "Weights", "Characters", "Weights", "Adverbs1", "Weights", "Adverbs2", "Weights", "Verbs", "Weights", "Prepositions", "Weights", "Adjectives1", "Weights", "Objects", "Weights", "Locations", "Weights"] 

# Prompting user to input public key and hint.
pubkey = input("Insert pubkey. ")
hint = input("Insert hint. ")

# Removing '0x' from beginning of public key and hint.
pubkey = pubkey[2:]
hint = hint[2:]

# Converting hint from hexadecimal to integer, then to binary form. Removing '0b' from beginning of binary hint and subsequently appended to variable 'result'. While the length of result is less than the desired 256-bits, a bit is appended from the beginning of the binary version of the public key. The result is 256 bits stored in 'result'.
def convertToBinary(pubkey, hint):
    result = ""
    hintbin = bin(int(hint, 16))
    hintbin = hintbin[2:]
    result += hintbin
    index = 0
    pubkeybin = bin(int(pubkey, 16))
    pubkeybin = pubkeybin[2:]
    while len(result) < 256:
      result += pubkeybin[index]
      index += 1
    return result

# Call binary conversion function for the inputted public key and hint.
bitString = convertToBinary(pubkey, hint)

results = []

# Iterate through binary 'result' from start value to index log base two of length of particular list in master list. Convert to an integer value. Append the value stored at this integer index within the particular list to an empty list 'results'. Append only weight values to empty list 'sum'. 
def getValues(start, end, i, decimal, results, sum):
    for el in bitString[start:end]:
        decimal = decimal*2 + int(el)
    results.append(lists[i][decimal])
    if listsNames[i] == "Weights":
      sum.append(float(lists[i][decimal]))
    
# First start value labelled index 0. First end value labelled index log base two of length of first list in master list.     
start = 0
i = 0
end = int(math.log2(len(lists[i])))
decimal = 0
sum = []

# Iterate through master list of nested lists, executing 'getValues' function and changing start value and end value upon each iteration. Start value becomes previous end value; end value becomes index log base two of length of following list in master list. Variable 'decimal' holding integer value is reset to zero.
while i < len(listsNames):
    getValues(start, end, i, decimal, results, sum)
    if i == len(listsNames) - 1:
      break
    else:
      start += int(math.log2(len(lists[i])))
      end += int(math.log2(len(lists[i + 1])))
      i += 1
      decimal = 0 

# Add all weight values yielded to variable 'test'. 
test = 0
for value in sum:
  test += value

# If sum of weights is positive, function proceeds to print values in a Midjourney-compatible fashion.If sum of weights is negative, Midjourney prompt will not proceed. In that case, all weights are flipped to be positive and values are printed in the same manner.
# A [detail][styles] with [colours][adjectives][adjectives][characters][adverbs][adverbs][verbs][prepositions][colours][adjectives][adjectives][objects] in [locations] --stylize x.
if test > 0:
  print(results[1] + " " + results[2] + "::" + results[3] + " " + results[4] + results[5] + results[6] + "::" + results[7] + " " + results[8] + "::" + results[9] + " " + results[10] + "::" + results[11] + " " + results[12] + "::" + results[13] + " " + results[14] + "::" + results[15] + " " + results[16] + "::" + results[17] + " " + results[18] + "::" + results[19] + " " + results[20] + "::" + results[21] + " " + results[22] + "::" + results[23] + " " + results[24] + "::" + results[25] + " " + results[26] + "::" + results[27] + " " + results[0])
  
else:
  new_list = []
  new_list = [el.replace("-", "") for el in results[1:]]
  new_list = [results[0]] + new_list
  print(new_list[1] + " " + new_list[2] + "::" + new_list[3] + " " + new_list[4] + new_list[5] + new_list[6] + "::" + new_list[7] + " " + new_list[8] + "::" + new_list[9] + " " + new_list[10] + "::" + new_list[11] + " " + new_list[12] + "::" + new_list[13] + " " + new_list[14] + "::" + new_list[15] + " " + new_list[16] + "::" + new_list[17] + " " + new_list[18] + "::" + new_list[19] + " " + new_list[20] + "::" + new_list[21] + " " + new_list[22] + "::" + new_list[23] + " " + new_list[24] + "::" + new_list[25] + " " + new_list[26] + "::" + new_list[27] + " " + new_list[0])
```
